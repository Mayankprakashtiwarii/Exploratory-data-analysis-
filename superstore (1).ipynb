```python
df["Product Base Margin"].fillna(df["Product Base Margin"].mean(),inplace=True)
```

```python
df['Order Priority'].value_counts()
```

```python
df['Ship Mode'].value_counts()
```

```python
df['Order Priority']=df['Order Priority'].replace("Critical ","Critical")
```

```python
pt.figure(figsize=(5,5))
pt.title("count of order priority")
sns.countplot(x="Order Priority",data=df,color='magenta')
pt.savefig("count of order priority.jpg")
pt.show()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
pt.figure(figsize=(5,4))
pt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
pt.legend(loc=4)
pt.savefig("Ship Mode.jpg")
pt.show()
```
